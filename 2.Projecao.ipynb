# Modelos de Projeção

Nesse script irei gerar projeçoes para os anos de 2023 e 2024 em termos de:

- Volume de contratos;
- Valor bruto.
    
Os dados estao agregados por ano e mês, pois na análise exploratória foi visto que nao há indícios de sazonalidade diária. 

Para fazer as projeçoes irei utilizar o algoritmo prophet da Meta, pois se trata de um método robusto de forecast em séries nao estacionárias e com outliers e grandes mudanças de tendências. 

Para organizar melhor o script, criei uma classe com os métodos necessários para ajustar e tunar o modelo. Essa classe está no script `Forecast_PB.py`

- Leitura dos dados pré agregados por ano e mês:

```python
df=pd.read_csv("dados/tb_modelo.csv")
```

```python
df
```

### Modelo de projeção - volume de contratos:

```python
df_model_contrato=(df[['anomes','qtd_contratos']]
                   .rename(columns={'anomes':'ds',
                                    'qtd_contratos':'y'})
                  ) #Selecionando coluna e renomeando para o prophet
```

```python
df_model_contrato
```

```python
forecast_contrato=ForecastPB(df_model_contrato) #instanciando a classe do modelo 
```

```python
forecast_contrato.tunar_modelo() #encontrando os melhores parametros para o modelo
```

Usando o crossvalidation com 90 dias de prediçao, esse modelo teve um erro percentual médio (mape) em torno de 8,28%:

```python
forecast_contrato.df_p  
```

```python
forecast_contrato.melhores_parametros
```

Pelo modelo temos uma tendência de crescimento na quantidade de contratos porém há um aumento de incerteza na prediçao conforme o horizonte amplia (vide aumento da amplitude dos intervalos de prediçao), pois estamos utilizando apenas 4 anos de dados para prever dois anos futuros:

```python
forecast_contrato.predict(meses=25)
```

Tabela com a previsao de contratos:

```python
forecast_contrato.forecast[['ds','yhat','yhat_lower','yhat_upper']].tail(25).rename(columns={'ds':'data',
                                                                                            'yhat':'qtd_contratos',
                                                                                            'yhat_lower':'intervalo_inferior',
                                                                                            'yhat_upper':'intervalo_superior'})
```

### Modelo de projeção - valor bruto:

```python
df_model_valor=(df[['anomes','sum_vlr_bruto']]
                .rename(columns={'anomes':'ds',
                                 'sum_vlr_bruto':'y'})
               )#Selecionando coluna e renomeando para o prophet
```

```python
forecast_valor=ForecastPB(df_model_valor) #instanciando a classe do modelo 
```

```python
forecast_valor.tunar_modelo() #encontrando os melhores parametros para o modelo
```

Usando o crossvalidation com 90 dias de prediçao, esse modelo teve um erro percentual médio (mape) em torno de 11,39%:

```python
forecast_valor.df_p
```

```python
forecast_valor.melhores_parametros
```

Pelo modelo temos uma tendência de crescimento mais acentuada em valor bruto porém também há um aumento de incerteza na prediçao conforme o horizonte amplia pelos mesmos motivos já comentados anteriormente:

```python
forecast_valor.predict(meses=25)
```

Tabela com a previsao de valor bruto:

```python
forecast_valor.forecast[['ds','yhat','yhat_lower','yhat_upper']].tail(25).rename(columns={'ds':'data',
                                                                                            'yhat':'valor',
                                                                                            'yhat_lower':'intervalo_inferior',
                                                                                            'yhat_upper':'intervalo_superior'}).round(0)
```
